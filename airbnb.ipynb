```python
import pandas as pd

df = pd.read_csv(
    "Listings.csv",
    encoding="latin1",
    engine="python",
    sep=",",
    quotechar='"',
    on_bad_lines="skip"
)
```

```python
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
```

```python
df.info()
```

```python
filtered_df = df[
    df['city'] == 'Paris'
][['host_since', 'neighbourhood', 'city', 'accommodates', 'price']]

filtered_df.head()
```

```python
filtered_df.isna().sum()
```

```python
(filtered_df.isna().sum() / len(filtered_df)) * 100
```

```python
numeric_df = filtered_df.select_dtypes(include=['number'])
```

```python
numeric_df.describe()
```

```python
paris_listings_neighbourhood = (
    filtered_df
    .groupby('neighbourhood')['price']
    .mean()
    .reset_index()
    .sort_values(by='price', ascending=True)
)
```

```python
paris_listings_neighbourhood.head()
```

```python
most_expensive_neighbourhood = (
    paris_listings_neighbourhood
    .sort_values(by='avg_price', ascending=False)
    .iloc[0]['neighbourhood']
)
```

```python
expensive_area_df = filtered_df[
    filtered_df['neighbourhood'] == most_expensive_neighbourhood
]
```

```python
paris_listings_accomodations = (
    expensive_area_df
    .groupby('accommodates')['price']
    .mean()
    .reset_index()
    .sort_values(by='price', ascending=True)
)
```

```python
paris_listings_accomodations.head()
```

```python
paris_listings_accomodations = paris_listings_accomodations.rename(
    columns={'price': 'avg_price'}
)
```

```python
filtered_df['host_year'] = filtered_df['host_since'].dt.year
```

```python
paris_listings_over_time = (
    filtered_df
    .groupby('host_year')
    .agg(
        avg_price=('price', 'mean'),
        new_hosts=('host_year', 'count')
    )
    .reset_index()
    .sort_values(by='host_year')
)
```

```python
paris_listings_over_time.head()
```

```python
plt.figure(figsize=(10, 8))

plt.barh(
    paris_listings_neighbourhood['neighbourhood'],
    paris_listings_neighbourhood['avg_price']
)

plt.xlabel("Average Price (€)")
plt.ylabel("Neighbourhood")
plt.title("Average Airbnb Price by Neighbourhood in Paris")

plt.tight_layout()
plt.show()
```

```python
top10 = paris_listings_neighbourhood.tail(10)

plt.figure(figsize=(10, 6))
plt.barh(top10['neighbourhood'], top10['avg_price'])

plt.xlabel("Average Price (€)")
plt.ylabel("Neighbourhood")
plt.title("Top 10 Most Expensive Neighbourhoods in Paris")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.barh(
    paris_listings_accomodations['accommodates'],
    paris_listings_accomodations['avg_price']
)

plt.xlabel("Average Price (€)")
plt.ylabel("Number of People Accommodated")
plt.title("Average Price by Accommodation Size in Paris' Most Expensive Neighbourhood")

plt.tight_layout()
plt.show()
```

```python
plt.figure()

plt.plot(
    paris_listings_over_time['host_year'],
    paris_listings_over_time['new_hosts']
)

plt.ylim(bottom=0)  # y-axis starts at 0
plt.xlabel("Year")
plt.ylabel("Number of New Hosts")
plt.title("Number of New Airbnb Hosts Over Time in Paris")

plt.show()
```

```python
plt.plot(
    paris_listings_over_time['host_year'],
    paris_listings_over_time['avg_price']
)

plt.ylim(bottom=0)  # y-axis starts at 0
plt.xlabel("host_year")
plt.ylabel("Average Price (€)")
plt.title("Average Airbnb Price Over Time in Paris")

plt.show()
```

```python
fig, ax1 = plt.subplots()

# Left axis → New Hosts
ax1.plot(
    paris_listings_over_time['host_year'],
    paris_listings_over_time['new_hosts']
)
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of New Hosts")
ax1.set_ylim(bottom=0)

# Right axis → Average Price
ax2 = ax1.twinx()
ax2.plot(
    paris_listings_over_time['host_year'],
    paris_listings_over_time['avg_price']
)
ax2.set_ylabel("Average Price (€)")
ax2.set_ylim(bottom=0)

# Title
plt.title("New Airbnb Hosts and Average Price Over Time in Paris")

plt.show()
```
